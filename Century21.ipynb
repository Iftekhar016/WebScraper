```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas

#An older website is being used due to introductory experience purposes.
headers = {'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'}
r = rq.get("http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", headers=headers)

c = r.content
soup = bs(c, "html.parser")

all = soup.find_all("div", {"class":"propertyRow"})
```

```python
itemsList =[]
for item in all:
    d ={}
    d["Price"] = item.find("h4", {"class":"propPrice"}).text.replace("\n","").replace(" ","")
    d["Street Address"] = item.find_all("span", {"class":"propAddressCollapse"})[0].text
    d["Locality"] = item.find_all("span", {"class":"propAddressCollapse"})[1].text
    
    try:
        d["Beds"] = item.find("span", {"class":"infoBed"}).find("b").text
    except:
        d["Beds"] = "None"
        
    try:
        d["sqft"] = item.find("span", {"class": "infoSqFt"}).find("b").text
    except:
        d["sqft"] = "None"
        
    try:
        d["fullBaths"] = item.find("span", {"class": "infoValueFullBath"}).find("b".text)
    except:
        d["fullBaths"] = "None"
    
    try:
        d["halfBath"] = item.find("span", {"class": "infoValueHalfBath"}).find("b".text)
    except:
        d["halfBath"] = "None"
    
    for column in item.find_all("div", {"class":"columnGroup"}):
        
        for featGroup, featName in zip(column.find_all("span", {"class":"featureGroup"}), column.find_all("span", {"class":"featureName"})):
            if "Lot Size" in featGroup.text:
                d["Lot Size"] = featName.text.replace(",", "")
    
    itemsList.append(d)
#     print(f"Price: [price]")
#     print(f"Address: [strtAdd] [cityAdd]")
#     print(f"Beds: [beds]")
#     print(f"Square Feet: [sqft]")
#     print(f"Full Baths: [fullBath]")
#     print(f"Half Baths: [halfBath]")
#     print(" ")
```

```python
df = pandas.DataFrame(itemsList)
df
```
